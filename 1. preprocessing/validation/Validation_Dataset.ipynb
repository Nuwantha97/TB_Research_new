# Validation Dataset : GSE19439_series_matrix

```python
# Set the file path
file_path = r'..\..\..\Datasets\validation\raw\GSE19439_series_matrix.txt'
```

```python
# Number of rows to be skipped (updated for this dataset)
m = 25

# Load the TXT file into a Pandas DataFrame
import pandas as pd

# Skip the header lines to get to the actual data
validation_set = pd.read_csv(file_path, sep='\t', skiprows=m)
```

```python
# Display the first 5 rows
validation_set.head()
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_validation_set = validation_set.T

# Reset the index to make the transposed column names into a header
transpose_validation_set.columns = transpose_validation_set.iloc[0]

# Drop the first row (index 0) which has the headers duplicated
transpose_validation_set = transpose_validation_set.drop(transpose_validation_set.index[0])

# Display the first few rows of the corrected transposed data
transpose_validation_set.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_validation_set.info()
```

## Handling Duplicate RowsHandling Duplicate Rows

```python
## Handling Duplicate Rows
transpose_validation_set.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

## Handling Duplicate Columns

```python
# Assign starting and end indices of duplicate columns to variables (updated for this dataset)
start_index = 8
end_index = 24

# Assign the indices related to the target columns (updated for this dataset)
tb_stage_index = 11  # Contains illness information (Latent, PTB, Control)
tb_type_index = 17   # Contains disease site information (pulm, extrapulm)

# Assign names to target columns
tb_stage_column = '!Sample_illness'
tb_type_column = '!Sample_disease_site'

# Rename the target columns
transpose_validation_set[tb_stage_column] = transpose_validation_set.iloc[:, tb_stage_index]
transpose_validation_set[tb_type_column] = transpose_validation_set.iloc[:, tb_type_index]

# Loop through the range of indices from start_index to end_index
for i in range(start_index, end_index + 1):
    # If the column index is one of our target columns, skip dropping it
    if i not in [tb_stage_index, tb_type_index]:
        # Drop the column
        transpose_validation_set.drop(transpose_validation_set.columns[i], axis=1, inplace=True)

# Reorder the columns - place the target columns at the front
new_column_order = [tb_stage_column, tb_type_column] + [col for col in transpose_validation_set.columns if col not in [tb_stage_column, tb_type_column]]
transpose_validation_set = transpose_validation_set[new_column_order]
```

```python
# Display the first few rows of the transposed data
transpose_validation_set.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_validation_set.columns[transpose_validation_set.nunique() == 1]

# Drop those columns
validation_data = transpose_validation_set.drop(columns=columns_to_drop)
```

```python
# Drop columns with more than 50% missing values
threshold = len(validation_data) * 0.5
validation_data = validation_data.dropna(thresh=threshold, axis=1)
```

```python
# Select columns that start with 'ILMN_' or are the target columns
columns_to_keep = [tb_stage_column, tb_type_column] + [col for col in validation_data.columns if col.startswith('ILMN_')]

# Filter the DataFrame to keep only the selected columns
validation_data = validation_data[columns_to_keep]
```

```python
# Select the first 5 columns
first_5_columns = validation_data.columns[:5]

# Create an empty list to store the rows
rows = []

# Loop through each of the first 5 columns
for col in first_5_columns:
    # Get the data type of the column using the Series dtype
    data_type = validation_data[col].dtypes

    # Get the first two row values of the column
    row_value_1 = validation_data[col].iloc[0]
    row_value_2 = validation_data[col].iloc[1]

    # Append the information as a row to the list
    rows.append({
        'Feature': col,
        'Data Type': data_type,
        'Row Value 1': row_value_1,
        'Row Value 2': row_value_2
    })

# Convert the list of rows into a DataFrame
result_df = pd.DataFrame(rows)

# Display the result
result_df
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the illness and disease site columns
unique_illness = validation_data[tb_stage_column].unique()
unique_disease_site = validation_data[tb_type_column].unique()

print("Unique illness categories:")
print(unique_illness)
print("\nUnique disease site categories:")
print(unique_disease_site)
```

```python
# Define functions to categorize each sample type
def categorize_tb_status(illness_value, disease_site_value):
    illness_value = str(illness_value).lower()
    
    if 'ptb' in illness_value or 'latent' in illness_value:
        return 'TB'
    elif 'control (bcg+)' in illness_value or 'control (bcg-)' in illness_value:
        return 'Healthy Control'
    else:
        return 'Other Disease'

def categorize_tb_stage(illness_value, disease_site_value):
    illness_value = str(illness_value).lower()
    
    if 'ptb' in illness_value:
        return 'Active TB'
    elif 'latent' in illness_value:
        return 'Latent TB'
    else:
        return 'N/A'  # For healthy controls and other diseases

def categorize_tb_type(illness_value, disease_site_value):
    illness_value = str(illness_value).lower()
    disease_site_value = str(disease_site_value).lower()
    
    # For PTB cases, always Pulmonary TB
    if 'ptb' in illness_value:
        return 'Pulmonary TB'
    # For Latent cases, check the disease site
    elif 'latent' in illness_value:
        if 'pulm' in disease_site_value:
            return 'Pulmonary TB'
        else:
            return 'Extra Pulmonary TB'
    else:
        return 'N/A'  # For healthy controls and other diseases
```

```python
# Apply the functions to create the three target columns
validation_data['TB_Status'] = validation_data.apply(
    lambda row: categorize_tb_status(row[tb_stage_column], row[tb_type_column]), axis=1
)

validation_data['TB_Stage'] = validation_data.apply(
    lambda row: categorize_tb_stage(row[tb_stage_column], row[tb_type_column]), axis=1
)

validation_data['TB_Type'] = validation_data.apply(
    lambda row: categorize_tb_type(row[tb_stage_column], row[tb_type_column]), axis=1
)
```

```python
# Display unique values in the new target columns to verify
print("TB_Status unique values:")
print(validation_data['TB_Status'].unique())

print("\nTB_Stage unique values:")
print(validation_data['TB_Stage'].unique())

print("\nTB_Type unique values:")
print(validation_data['TB_Type'].unique())
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = validation_data.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    validation_data[col] = cap_outliers(validation_data[col])

# Display the random 5 rows to verify the changes
validation_data.sample(5)
```
