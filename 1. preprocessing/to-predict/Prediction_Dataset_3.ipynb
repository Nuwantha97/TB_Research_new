# Prediction Dataset 3 : GSE42825_series_matrix

```python
# Set the file path
file_path = r'..\..\..\Datasets\to-predict\raw\GSE42825_series_matrix.txt'
```

```python
# Number of rows to be skipped
m = 50

# Load the TXT file into a Pandas DataFrame
import pandas as pd

# Skip the header lines (e.g., metadata) to get to the actual data
test_set = pd.read_csv(file_path, sep='\t', skiprows=m)
```

```python
# Display the first 5 rows
test_set.head()
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_test_set = test_set.T

# Reset the index to make the transposed column names into a header
transpose_test_set.columns = transpose_test_set.iloc[0]

# Drop the first row (index 0) which has the headers duplicated
transpose_test_set = transpose_test_set.drop(transpose_test_set.index[0])

# Display the first few rows of the corrected transposed data
transpose_test_set.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_test_set.info()
```

## Handling Duplicate RowsHandling Duplicate Rows

```python
transpose_test_set.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

## Handling Duplicate Columns

```python
# Assign starting and end indices of duplicate columns to variables
start_index = 8
end_index = 11

# Assign the index related to the target column
target_column_index = 10

# Assign a name to target column
target_column ='!Sample_disease_state'

# Rename the target column (rather than dropping it)
transpose_test_set[target_column] = transpose_test_set.iloc[:, target_column_index]

# Loop through the range of indices from start_index to end_index
for i in range(start_index, end_index + 1):
    # If the column index is the target column, skip renaming it
    if i != target_column_index:
        # Drop the column
        transpose_test_set.drop(transpose_test_set.columns[i], axis=1, inplace=True)

# Reorder the columns
# Here, place the target column at the front
new_column_order = [target_column] + [col for col in transpose_test_set.columns if col != target_column]
transpose_test_set = transpose_test_set[new_column_order]
```

```python
# Display the first few rows of the transposed data
transpose_test_set.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_test_set.columns[transpose_test_set.nunique() == 1]

# Drop those columns
test_data = transpose_test_set.drop(columns=columns_to_drop)
```

```python
# Drop columns with more than 50% missing values
# Calculate threshold for dropping columns
threshold = len(test_data) * 0.5
test_data = test_data.dropna(thresh=threshold, axis=1)
```

```python
# Select columns that start with 'ILMN_' or are the target column
columns_to_keep = [target_column] + [col for col in test_data.columns if col.startswith('ILMN_')]

# Filter the DataFrame to keep only the selected columns
test_data = test_data[columns_to_keep]
```

```python
# Select the first 5 columns
first_5_columns = test_data.columns[:5]

# Create an empty list to store the rows
rows = []

# Loop through each of the first 5 columns
for col in first_5_columns:
    # Get the data type of the column using the Series dtype
    data_type = test_data[col].dtypes  # Access the dtype via .dtypes (not .dtype)

    # Get the first two row values of the column
    row_value_1 = test_data[col].iloc[0]
    row_value_2 = test_data[col].iloc[1]

    # Append the information as a row to the list
    rows.append({
        'Feature': col,
        'Data Type': data_type,
        'Row Value 1': row_value_1,
        'Row Value 2': row_value_2
    })

# Convert the list of rows into a DataFrame
result_df = pd.DataFrame(rows)

# Display the result
result_df
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the "!Sample_disease_state" column
unique_categories = test_data[target_column].unique()

# Display the unique categories
unique_categories
```

```python
# Define a function to categorize each sample type
def categorize_sample_type(source_name):
    source_name = source_name.lower()
    if 'disease state: tb' in source_name:
        return 'TB'
    elif 'disease state: control' in source_name:
        return 'Healthy Control'
    else:
        return 'Other Disease'  # Other lung disease like 'Pneumonia'

# Apply the function to the "!Sample_disease_state" column
test_data['TB_Status'] = test_data[target_column].apply(categorize_sample_type)

# Display unique values in the new 'TB_Status' column to verify
print(test_data['TB_Status'].unique())
```

```python
# Drop the '!Sample_disease_state' column.
test_data.drop(target_column, axis=1, inplace=True)
```

## Data Type Conversion

```python
# Convert columns that should be numeric (e.g., ILMN_1343291, ILMN_1343295)
# Select columns that start with "ILMN_"
ilmn_columns = test_data.columns[test_data.columns.str.startswith('ILMN_')]

# Convert all 'ILMN_' columns to numeric
test_data[ilmn_columns] = test_data[ilmn_columns].apply(pd.to_numeric, errors='coerce')

# Verify the data types of the columns
print(test_data[ilmn_columns].dtypes)
```

```python
# Convert target column ('TB_Status') to categorical
test_data['TB_Status'] = test_data['TB_Status'].astype('category')

# Verify the data type of the column
print(test_data['TB_Status'].dtypes)
```

## Standardizing/Normalizing Numeric Data

```python
from sklearn.preprocessing import StandardScaler

# Select only the numeric columns in the dataset
numeric_columns = test_data.select_dtypes(include=['float64', 'int64']).columns

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the numeric columns and transform them
test_data[numeric_columns] = scaler.fit_transform(test_data[numeric_columns])

# Display random 5 rows to verify the standardization
test_data.sample(5)
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = test_data.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    test_data[col] = cap_outliers(test_data[col])

# Display the random 5 rows to verify the changes
test_data.sample(5)
```
