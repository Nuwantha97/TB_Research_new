# Dataset 1 : GSE19491_series_matrix

```python
# Import necessary library
import pandas as pd

# Set the file path for the input TXT file
file_path = r"/content/drive/MyDrive/Research/Dataset01-GSE19491/GSE19491_series_matrix.txt"

# Load the TXT file into a Pandas DataFrame, skipping metadata lines
dataset1 = pd.read_csv(file_path, sep='\t', skiprows=33, low_memory=False)
```

```python
# Display the first 5 rows
dataset1.head()
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_dataset1 = dataset1.T

# Reset the index to make the transposed column names into a header
transpose_dataset1.columns = transpose_dataset1.iloc[0]
transpose_dataset1 = transpose_dataset1.drop(transpose_dataset1.index[0])

# Display the first few rows of the transposed data
transpose_dataset1.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_dataset1.info()
```

## Handling Duplicate Rows

```python
transpose_dataset1.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

## Handling Duplicate Columns

```python
# Identify columns that start with '!Sample_characteristics_ch1' by their index positions
# The duplicate columns are in positions known by inspection with indices 8, 9, 10, and 11
characteristics_indices = [8, 9, 10, 11]

# Extract and rename each column using its index position
transpose_dataset1['!Sample_age'] = transpose_dataset1.iloc[:, characteristics_indices[0]]
transpose_dataset1['!Sample_gender'] = transpose_dataset1.iloc[:, characteristics_indices[1]]
transpose_dataset1['!Sample_ethnicity'] = transpose_dataset1.iloc[:, characteristics_indices[2]]
transpose_dataset1['!Sample_illness'] = transpose_dataset1.iloc[:, characteristics_indices[3]]

# Drop the original duplicated columns by their indices
transpose_dataset1.drop(transpose_dataset1.columns[characteristics_indices], axis=1, inplace=True)

# Reorder the columns to place the new columns at the front
# Create a new order by specifying the newly created columns first
new_column_order = ['!Sample_age', '!Sample_gender', '!Sample_ethnicity', '!Sample_illness'] + \
                   [col for col in transpose_dataset1.columns if col not in ['!Sample_age', '!Sample_gender', '!Sample_ethnicity', '!Sample_illness']]

# Reassign columns to the DataFrame in the new order
transpose_dataset1 = transpose_dataset1[new_column_order]

```

```python
# Display the first few rows of the transposed data
transpose_dataset1.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_dataset1.columns[transpose_dataset1.nunique() == 1]

# Drop those columns
data1 = transpose_dataset1.drop(columns=columns_to_drop)
```

```python
# Drop columns with more than 50% missing values
# Calculate threshold for dropping columns
threshold = len(data1) * 0.5
data1 = data1.dropna(thresh=threshold, axis=1)
```

```python
# Drop the columns which won’t contribute to distinguishing TB states.
columns_to_drop = [
    '!Sample_geo_accession', '!Sample_submission_date', '!Sample_last_update_date',
    '!Sample_extract_protocol_ch1', '!Sample_label_protocol_ch1', '!Sample_hyb_protocol',
    '!Sample_description', '!Sample_data_processing', '!Sample_supplementary_file',
    '!Sample_data_row_count','!Sample_age','!Sample_gender','!Sample_ethnicity',
    '!Sample_source_name_ch1','ID_REF'
]

data1.drop(columns=columns_to_drop, inplace=True)
```

```python
# Select the first 5 columns
first_5_columns = data1.columns[:5]

# Create an empty list to store the rows
rows = []

# Loop through each of the first 5 columns
for col in first_5_columns:
    # Get the data type of the column using the Series dtype
    data_type = data1[col].dtypes  # Access the dtype via .dtypes (not .dtype)

    # Get the first two row values of the column
    row_value_1 = data1[col].iloc[0]
    row_value_2 = data1[col].iloc[1]

    # Append the information as a row to the list
    rows.append({
        'Feature': col,
        'Data Type': data_type,
        'Row Value 1': row_value_1,
        'Row Value 2': row_value_2
    })

# Convert the list of rows into a DataFrame
result_df = pd.DataFrame(rows)

# Display the result
result_df
```

## Handling Missing values

```python
# Check for missing values
data1.isnull().sum()
```

```python
# Get columns with missing values and their data types
missing_values = data1.isnull().sum()
missing_columns = missing_values[missing_values > 0].index
missing_data_types = data1[missing_columns].dtypes

# Combine column names, missing values count, and data types into a DataFrame
missing_summary = pd.DataFrame({
    'Column Name': missing_columns,
    'Missing Values Count': missing_values[missing_columns],
    'Data Type': missing_data_types
}).reset_index(drop=True)

print(missing_summary)
```

```python
# Identify columns with missing values
columns_with_missing_values = data1.columns[data1.isnull().any()]

# Convert the identified columns to numeric type, coercing errors if necessary
data1[columns_with_missing_values] = data1[columns_with_missing_values].apply(pd.to_numeric, errors='coerce')

# Fill missing values with the mean of each column
data1[columns_with_missing_values] = data1[columns_with_missing_values].fillna(data1[columns_with_missing_values].mean())
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the "!Sample_illness" column
unique_categories = data1["!Sample_illness"].unique()

# Display the unique categories
unique_categories
```

```python
# Define a function to categorize each sample type
def categorize_sample_type(source_name):
    source_name = source_name.lower()
    if 'ptb' in source_name:
        return 'Active TB'
    elif 'latent' in source_name:
        return 'Latent TB'
    elif 'control' in source_name:
        return 'Healthy Control'
    else:
        return 'Other Disease'

# Apply the function to the "!Sample_illness" column
data1['TB_Status'] = data1['!Sample_illness'].apply(categorize_sample_type)

# Display unique values in the new 'TB_Status' column to verify
print(data1['TB_Status'].unique())
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = data1.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    data1[col] = cap_outliers(data1[col])

# Display the random 5 rows to verify the changes
data1.sample(5)
```
