# Dataset 5 : GSE152532_series_matrix

```python
# Import necessary library
import pandas as pd

# Set the file path for the input TXT file
file_path = r"..\..\..\Datasets\train\raw\GSE152532_series_matrix.txt"

# Load the TXT file into a Pandas DataFrame, skipping metadata lines
dataset5 = pd.read_csv(file_path, sep='\t', skiprows=28)
```

```python
# Display the first 5 rows
dataset5.head()
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_dataset5 = dataset5.T

# Reset the index to make the transposed column names into a header
transpose_dataset5.columns = transpose_dataset5.iloc[0]
transpose_dataset5 = transpose_dataset5.drop(transpose_dataset5.index[0])

# Display the first few rows of the transposed data
transpose_dataset5.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_dataset5.info()
```

## Handling Duplicate Rows

```python
transpose_dataset5.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

## Handling Duplicate Columns

```python
# Identify columns that start with '!Sample_characteristics_ch1' by their index positions
# The duplicate columns are in positions known by inspection with indices 8, 9, 10, and 11
characteristics_indices = [9]

# Extract and rename each column using its index position
transpose_dataset5['!Sample_stage'] = transpose_dataset5.iloc[:, characteristics_indices[0]]

# Drop the original duplicated columns by their indices
transpose_dataset5.drop(transpose_dataset5.columns[characteristics_indices], axis=1, inplace=True)

# Reorder the columns to place the new columns at the front
# Create a new order by specifying the newly created columns first
new_column_order = ['!Sample_stage'] + \
                   [col for col in transpose_dataset5.columns if col not in ['!Sample_stage']]

# Reassign columns to the DataFrame in the new order
transpose_dataset5 = transpose_dataset5[new_column_order]

```

```python
# Display the first few rows of the transposed data
transpose_dataset5.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_dataset5.columns[transpose_dataset5.nunique() == 1]

# Drop those columns
data5 = transpose_dataset5.drop(columns=columns_to_drop)
```

```python
# Drop columns with more than 50% missing values
# Calculate threshold for dropping columns
threshold = len(data5) * 0.5
data5 = data5.dropna(thresh=threshold, axis=1)
```

```python
# Drop the columns which won’t contribute to distinguishing TB states.
columns_to_drop = [
    '!Sample_geo_accession', '!Sample_supplementary_file', 'ID_REF'
]

data5.drop(columns=columns_to_drop, inplace=True)
```

```python
# Select the first 5 columns
first_5_columns = data5.columns[:5]

# Create an empty list to store the rows
rows = []

# Loop through each of the first 5 columns
for col in first_5_columns:
    # Get the data type of the column using the Series dtype
    data_type = data5[col].dtypes  # Access the dtype via .dtypes (not .dtype)

    # Get the first two row values of the column
    row_value_1 = data5[col].iloc[0]
    row_value_2 = data5[col].iloc[1]

    # Append the information as a row to the list
    rows.append({
        'Feature': col,
        'Data Type': data_type,
        'Row Value 1': row_value_1,
        'Row Value 2': row_value_2
    })

# Convert the list of rows into a DataFrame
result_df = pd.DataFrame(rows)

# Display the result
result_df
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the "!Sample_illness" column
unique_categories = data5["!Sample_stage"].unique()

# Display the unique categories
unique_categories
```

```python
# Define a function to categorize each sample type
def categorize_sample_type(source_name):
    source_name = source_name.lower()
    if 'cohort: active tb' in source_name:
        return 'Active TB'
    elif 'cohort: latent tb' in source_name:
        return 'Latent TB'
    else:
        return 'Healthy Control'

# Apply the function to the "!Sample_illness" column
data5['TB_Status'] = data5['!Sample_stage'].apply(categorize_sample_type)

# Display unique values in the new 'TB_Status' column to verify
print(data5['TB_Status'].unique())
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = data5.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    data5[col] = cap_outliers(data5[col])

# Display the random 5 rows to verify the changes
data5.sample(5)
```
