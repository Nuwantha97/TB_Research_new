# Dataset 2 : GSE42830_series_matrix

```python
# Import necessary library
import pandas as pd

# Set the file path for the input TXT file
file_path = r"/content/drive/MyDrive/Research/Dataset02-GSE42830/GSE42830_series_matrix.txt"

# Load the TXT file into a Pandas DataFrame, skipping metadata lines
dataset2 = pd.read_csv(file_path, sep='\t', skiprows=50)
```

```python
# Display the first 5 rows
dataset2.head()
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_dataset2 = dataset2.T

# Reset the index to make the transposed column names into a header
transpose_dataset2.columns = transpose_dataset2.iloc[0]
transpose_dataset2 = transpose_dataset2.drop(transpose_dataset2.index[0])

# Display the first few rows of the transposed data
transpose_dataset2.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_dataset2.info()
```

## Handling Duplicate Rows

```python
transpose_dataset2.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

### Handling Duplicate Columns

```python
# Identify columns that start with '!Sample_characteristics_ch1' by their index positions
# The duplicate columns are in positions known by inspection with indices 8, 9, 10, and 11
characteristics_indices = [8, 9, 10, 11]

# Extract and rename each column using its index position
transpose_dataset2['!Sample_gender'] = transpose_dataset2.iloc[:, characteristics_indices[0]]
transpose_dataset2['!Sample_ethnicity'] = transpose_dataset2.iloc[:, characteristics_indices[1]]
transpose_dataset2['!Sample_disease_state'] = transpose_dataset2.iloc[:, characteristics_indices[2]]
transpose_dataset2['!Sample_tissue'] = transpose_dataset2.iloc[:, characteristics_indices[3]]

# Drop the original duplicated columns by their indices
transpose_dataset2.drop(transpose_dataset2.columns[characteristics_indices], axis=1, inplace=True)

# Reorder the columns to place the new columns at the front
# Create a new order by specifying the newly created columns first
new_column_order = ['!Sample_gender', '!Sample_ethnicity', '!Sample_disease_state', '!Sample_tissue'] + \
                   [col for col in transpose_dataset2.columns if col not in ['!Sample_gender', '!Sample_ethnicity', '!Sample_disease_state', '!Sample_tissue']]

# Reassign columns to the DataFrame in the new order
transpose_dataset2 = transpose_dataset2[new_column_order]

```

```python
# Display the first few rows of the transposed data
transpose_dataset2.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_dataset2.columns[transpose_dataset2.nunique() == 1]

# Drop those columns
data2 = transpose_dataset2.drop(columns=columns_to_drop)
```

```python
# Drop columns with more than 50% missing values
# Calculate threshold for dropping columns
threshold = len(data2) * 0.5
data2 = data2.dropna(thresh=threshold, axis=1)
```

```python
# Drop the columns which won’t contribute to distinguishing TB states.
columns_to_drop = [
    '!Sample_geo_accession','!Sample_gender','!Sample_ethnicity','ID_REF'
]

data2.drop(columns=columns_to_drop, inplace=True)
```

```python
# Select the first 5 columns
first_5_columns = data2.columns[:5]

# Create an empty list to store the rows
rows = []

# Loop through each of the first 5 columns
for col in first_5_columns:
    # Get the data type of the column using the Series dtype
    data_type = data2[col].dtypes  # Access the dtype via .dtypes (not .dtype)

    # Get the first two row values of the column
    row_value_1 = data2[col].iloc[0]
    row_value_2 = data2[col].iloc[1]

    # Append the information as a row to the list
    rows.append({
        'Feature': col,
        'Data Type': data_type,
        'Row Value 1': row_value_1,
        'Row Value 2': row_value_2
    })

# Convert the list of rows into a DataFrame
result_df = pd.DataFrame(rows)

# Display the result
result_df
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the "!Sample_disease_state" column
unique_categories = data2["!Sample_disease_state"].unique()

# Display the unique categories
unique_categories
```

```python
# Define a function to categorize each sample type
def categorize_sample_type(source_name):
    source_name = source_name.lower()
    if 'tb' in source_name:
        return 'TB'
    elif 'control' in source_name:
        return 'Healthy Control'
    else:
        return 'Other Disease'  # Other lung disease like 'Lung Cancer'

# Apply the function to the "!Sample_disease_state" column
data2['TB_Status'] = data2['!Sample_disease_state'].apply(categorize_sample_type)

# Display unique values in the new 'TB_Status' column to verify
print(data2['TB_Status'].unique())
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = data2.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    data2[col] = cap_outliers(data2[col])

# Display the random 5 rows to verify the changes
data2.sample(5)
```
